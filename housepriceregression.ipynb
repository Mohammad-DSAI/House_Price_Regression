```python
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
import pandas as pd
train_df = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/train.csv')
print(train_df.shape)
train_df.head()
```

```python
test_df = pd.read_csv('/kaggle/input/house-prices-advanced-regression-techniques/test.csv')
print(test_df.shape)
test_df.head()
```

```python
train_df['SalePrice'].mean()
```

```python
test_df['SalePrice'] = pd.Series([train_df['SalePrice'].mean()]*len(train_df))
test_df
```

```python
test_df[['Id','SalePrice']].to_csv('Average_Prediction.csv', index = False)
```

```python
import copy
train_df_sub = copy.deepcopy(train_df[['MSSubClass', 'LotFrontage', 'SalePrice']])
train_df_sub = train_df_sub.fillna(train_df_sub.mean())
train_df_sub
```

```python
x_train, y_train = train_df_sub.to_numpy()[:,:-1], train_df_sub.to_numpy()[:,-1]
y_train
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(x_train, y_train)
lr
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
train_df_sub['linear-prediction'] = lr.predict(x_train)
train_df_sub
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(train_df_sub['SalePrice'], train_df_sub['linear-prediction'])
```

```python
test_df_sub = copy.deepcopy(test_df[['MSSubClass', 'LotFrontage']])
test_df_sub
```

```python
test_df_sub = test_df_sub.fillna(test_df_sub.mean())
test_df_sub
```

```python
x_test = test_df_sub.to_numpy()
x_test
```

```python
test_df['SalePrice'] = lr.predict(x_test)
test_df
```

```python
test_df[['Id','SalePrice']].to_csv('Linear_Prediction.csv', index = False)
```
